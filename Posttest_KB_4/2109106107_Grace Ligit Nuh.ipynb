#### Data Cleaning

```python
ds = pd.read_csv('adult_copy.csv')
ds
```

Mendapatkan jumlah data null

```python
ds.isna().sum()
```

Membuat kolom age, educational-num, dan hours-per-week menjadi null

```python
# Mengambil indeks acak untuk beberapa baris dalam dataset
random_indexes_age = random.sample(list(df.index), 10)
random_indexes_educational_num = random.sample(list(df.index), 10)
random_indexes_hours_per_week = random.sample(list(df.index), 5)

# Mengubah nilai menjadi null untuk baris-baris dengan indeks yang dipilih secara acak
ds.loc[random_indexes_age, 'age'] = None
ds.loc[random_indexes_educational_num, 'educational-num'] = None
ds.loc[random_indexes_hours_per_week, 'hours-per-week'] = None
```

```python
ds.isna().sum()
```

```python
mean = SimpleImputer(strategy='mean')
median = SimpleImputer(strategy='median')
modus = SimpleImputer(strategy='most_frequent')
```

```python
ds['age'] = mean.fit_transform(ds[['age']])
ds['educational-num'] = median.fit_transform(ds[['educational-num']])
ds['hours-per-week'] = modus.fit_transform(ds[['hours-per-week']])
```

```python
ds
```

```python
ds.isna().sum()
```

Mendapatkan jumlah data duplicate

```python
ds.duplicated().sum()
```

```python
ds.drop_duplicates(inplace=True)
```

```python
print("Jumlah nilai Duplikat:", ds.duplicated().sum())
```
