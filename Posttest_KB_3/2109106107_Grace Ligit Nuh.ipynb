#### Import Library

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

#### Memanggil Dataset

```python
df = pd.read_csv('adult.csv')
df
```

#### Pie Chart
Pada pie chart di bawah ini, ditampilkan atribut pendapatan (income) yang memiliki dua data, yaitu "<=50k" dan ">50k," dengan "<=50k" memiliki jumlah yang lebih banyak dibandingkan dengan ">50k".

```python
df['income'].value_counts()
labels = df['income'].value_counts().rename_axis('income').reset_index(name='Jumlah')
labels
```

```python
plt.pie(df['income'].value_counts(), labels=labels['income'], autopct='%.1f%%', startangle=50)
plt.title("Jumlah Income")
plt.show()
```

#### Heatmap
Pada heatmap di bawah ini, ditampilkan korelasi antara atribut numerik dan pendapatan (income) untuk melihat hubungan atau keterkaitan antara atribut numerik dengan pendapatan.

Mengonversi "income" menjadi variabel numerik (1 untuk ">50K" dan 0 untuk "<=50K")

```python
df['income_numeric'] = df['income'].apply(lambda x: 1 if x.strip() == '>50K' else 0)
```

Memilih atribut numerik dan kolom "income_numeric" 

```python
atribut_numerik = ['age', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week', 'income_numeric']
```

Menghitung matriks korelasi

```python
correlation_matrix = df[atribut_numerik].corr()
```

Membuat heatmap korelasi

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Atribut Numerik dan Income')
plt.show()
```

#### Grafik Batang
Pada grafik batang di bawah ini, ditampilkan data pendapatan (income) dan jumlahnya untuk memvisualisasikan seberapa banyak jumlah pendapatan yang berada dalam kategori "<=50k" dan ">50k".

```python
jumlah_income = df['income'].value_counts()  
plt.figure(figsize=(10, 6))
jumlah_income.plot(kind='bar', color='lightgreen')
plt.title('Jumlah Income')
plt.xlabel('Income')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

#### Histogram
Pada histogram di bawah ini, ditampilkan data pendapatan (income) dan jumlahnya berdasarkan jenis kelamin, dengan tujuan untuk melihat seberapa banyak pendapatan yang terbagi berdasarkan jenis kelamin.

Memisahkan data berdasarkan jenis kelamin

```python
male_income = df[df['gender'] == 'Male']['income']
female_income = df[df['gender'] == 'Female']['income']
```

Membuat histogram untuk pendapatan berdasarkan jenis kelamin

```python

plt.hist([male_income, female_income], bins=5, color=['blue', 'pink'], label=['Male', 'Female'])
plt.title("Histogram Income berdasarkan Gender")
plt.xlabel('Jumlah Income')
plt.ylabel('Frekuensi Data')
plt.legend()  
plt.show()
```
